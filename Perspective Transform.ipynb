```python
plt.imshow(img)
plt.plot(60,415,".")
plt.plot(1450,415,".")
plt.plot(750,170,".")
plt.plot(840,170,".")
```

```python
def Warp(img):
    img_size = (img.shape[1],img.shape[0])
    src = np.float32([[750,170],[1450,415],[60,415],[640,170]])
    dst = np.float32([[1300,170],[1300,415],[200,415],[200,170]])
    M  = cv2.getPerspectiveTransform(src,dst)
    InvM = cv2.getPerspectiveTransform(dst,src)
    warped = cv2.warpPerspective(img,M,img_size,flags=cv2.INTER_LINEAR)
    return warped
```

```python
warped_img = Warp(img)
plt.imshow(warped_img)
```
