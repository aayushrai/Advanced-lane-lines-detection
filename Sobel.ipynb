```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("lane.png")
```

```python
plt.imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
sobelx = cv2.Sobel(gray,cv2.CV_64F,1,0)
sobely = cv2.Sobel(gray,cv2.CV_64F,0,1)
```

```python
plt.imshow(sobely,cmap="gray")
```

```python
abs_sobelx = np.absolute(sobelx)
scaled_sobelx = np.uint8(255*abs_sobelx/np.max(abs_sobelx))
plt.imshow(scaled_sobelx,cmap="gray")
```

```python
abs_sobely = np.absolute(sobely)
scaled_sobely = np.uint8(255*abs_sobely/np.max(abs_sobely))
plt.imshow(scaled_sobely,cmap="gray")
```

```python
thresh_min = 20
thresh_max = 100
sxbinary = np.zeros_like(scaled_sobel)
sxbinary[(scaled_sobel >= thresh_min) & (scaled_sobel <= thresh_max)] = 1
plt.imshow(sxbinary, cmap='gray')
```

```python
sxbinary2 = np.zeros_like(scaled_sobely)
sxbinary2[(scaled_sobely >= thresh_min) & (scaled_sobely <= thresh_max)] = 1
plt.imshow(sxbinary2, cmap='gray')
```
